```python
import requests
import pandas as pd
import csv 
import random
import numpy as np
```

```python
#Генерирование случайного id-адреса.
arr_id = [random.randint(100000000, 300000000) for _ in range(10000)]
#Преобразование массива случайных id во множество(без повторения элементов).
arr_id = set(arr_id)
print(arr_id)
print(len(arr_id))
```

```python
token = ''
version = 5.131
for i in range(len(arr_id)):
  user_id = arr_id[i]
  owner_id = arr_id[i]
  response = requests.get('https://api.vk.com/method/status.get', params={'access_token': token, 'user_id': user_id, 'v': version})
  photos = requests.get('https://api.vk.com/method/photos.getAll', params={'access_token': token, 'owner_id': owner_id, 'extended': 1, 'offset': 0, 'count': 20, 'photo_sizes': 0, 'no_service_albums': 0, 'need_hidden': 1, 'skip_hidden': 1,'v': version})
  data = response.json()
  data_photos = photos.json()
```
